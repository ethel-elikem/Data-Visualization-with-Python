## Univariate Exploration

> I'll start my visualizing the distribution of the main variable of interest; `LoanStatus` 




```python
## Plotting Function for Loan Status

def log_plt(dataset, x):
   
    fig = plt.figure(figsize = [20, 8])

    #  ensuring single coloring of bars
    base_color = sb.color_palette()[0]

    # ordering of bars
    freq = dataset[x].value_counts()
    status_order = freq.index
    
    # Add the main title
    fig.suptitle("Categories of Loan Status", fontsize=20)
    
    # HISTOGRAM ON LEFT: full data without scaling
    ax1 = plt.subplot(1, 2, 1)
    sb.countplot(data=dataset, x=x, color=base_color, order = status_order)
    plt.xticks(rotation=90)



    # HISTOGRAM ON LEFT: full data with scaling
    ax2 = plt.subplot(1, 2, 2)
    sb.countplot(data=dataset, x=x, color=base_color, order = status_order)
    plt.xticks(rotation=90)
    plt.yscale('log')
    # Add the text for each subplot
    ax1.title.set_text("Data without log scaling")
    ax2.title.set_text("Data with log scaling")

plt.show()
```

```python
log_plt(loan, 'LoanStatus')
```

> I'll proceed to visualize other variables of interest

```python
def hist_plt(d,d1, d2,d3, x):
    
    fig = plt.figure(figsize=[20, 10])
    
    # Add the main title
    fig.suptitle(x, fontsize=20)
    
    #  ensuring single coloring of bars
    base_color = sb.color_palette()[0]
    
    plt.subplot(2, 2, 1)
    plt.hist(data = d, x = x)
    plt.xlabel(x +' for all loans')


    plt.subplot(2, 2, 2)
    plt.hist(data = d1, x = x)
    plt.xlabel(x +' all loans except Current')


    plt.subplot(2, 2, 3)
    plt.hist(data = d2, x = x)
    plt.xlabel(x +' before 2009')


    plt.subplot(2, 2, 4)
    plt.hist(data = d3, x = x)
    plt.xlabel(x +' on or after July 2009')
plt.show()             
```

```python
#BorrowerAPR:
# The Borrower's Annual Percentage Rate (APR) for the loan.
hist_plt(loan, other_loans, before_2009, after_2009, 'BorrowerAPR')
```

```python
#BorrowerRate:
# The Borrower's interest rate for this loan. 
hist_plt(loan, other_loans, before_2009, after_2009, 'BorrowerRate')
```

> The distribution of the `BorrowerAPR` and `BorrowerRate` appear very similar and it is no surprise because they are related. In all four subplots the both appear to be unimodal and fairly normal except for their plots of the *before 2009* period, that slightly skewed to the left.
